# Backtester Demo Notebook

This notebook demonstrates usage of the `trading_backtester` package with real CSV data, running a couple of strategies, and plotting results.

```python
import pandas as pd
import matplotlib.pyplot as plt

# keep plots inline
%matplotlib inline
```

```python
# load sample data
csv_path = 'data/sample.csv'
data = pd.read_csv(csv_path, parse_dates=True, index_col=0)
data.head()
```

```python
from trading_backtester import BacktestEngine
from trading_backtester.strategies.moving_average import MovingAverageCrossover

# run a simple MA backtest
strategy = MovingAverageCrossover(short_window=2, long_window=3)
engine = BacktestEngine(data, strategy)
results = engine.run()
print(engine.summary())
```

```python
# plot equity curve
results['equity'] = (1 + results['strategy_returns']).cumprod()
results['equity'].plot(title='Equity Curve')
```

```python
from trading_backtester.strategies.mean_reversion import MeanReversion

strategy2 = MeanReversion(window=3, threshold=1)
engine2 = BacktestEngine(data, strategy2, transaction_cost=0.001)
res2 = engine2.run()
print(engine2.summary())
res2['equity'] = (1 + res2['strategy_returns']).cumprod()
res2['equity'].plot(title='Mean Rev Equity Curve')
```
